```python
def radix_sort(arr):
    # Find the maximum number in the array to determine the number of digits
    max_num = max(arr)
    num_digits = len(str(max_num))

    # Perform counting sort for each digit
    for i in range(num_digits):
        # Create a bucket for each digit (0-9)
        buckets = [[] for _ in range(10)]

        # Place each element in the corresponding digit bucket
        for num in arr:
            digit = (num // (10**i)) % 10
            buckets[digit].append(num)

        # Flatten the buckets and store the result back in the array
        arr = [num for bucket in buckets for num in bucket]

    return arr

arr = [12, 345, 67, 8, 901, 2345, 678, 123]
sorted_arr = radix_sort(arr)
print(sorted_arr)  # [8, 12, 67, 123, 345, 678, 901, 2345]
```

```python
with open("aoc_data/day3.txt", 'r') as f:
    data = f.read().splitlines()

print(data)
```

```python
def priority(x):
    if 'A' <= x <= 'Z':
        return ord(x) - ord('A') + 27
    elif 'a' <= x <= 'z':
        return ord(x) - ord('a') + 1

common = []
for line in data:
    first, second = line[:len(line) // 2], line[len(line) // 2:]
    common.append([x for x in first if x in second][0])

print(sum(priority(x) for x in common))
```

```python
# Split the data into chunks of 3 lines each
data = [data[i:i+3] for i in range(0, len(data), 3)]
badges = []
for group in data:
    common = set(group[0]) & set(group[1]) & set(group[2])
    badges.append(list(common)[0])

print(sum(priority(x) for x in badges))
```
